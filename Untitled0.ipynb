```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.core.fromnumeric import reshape
import pandas as pd
import datetime
import time
from sklearn import preprocessing
#Row-column value for csv file
ROWCOUNT = 8267068#35893#282#8267068
COLUMNCOUNT = 9

x_test =0
y_test =0
x_train =0
y_train =0
date =0
sec =0
high= 0.0
low= 0.0
close= 0.0
open = 0.0

datas      = pd.read_csv('/content/drive/MyDrive/Lisans Projesi/EURUSD_M1.csv', header=1)
#datas_test = pd.read_csv('EURUSD_M1_test.csv' , header=1)
#change value to list
datas      = datas.values.tolist()
#datas_test = datas_test.values.tolist()
#Split all columns and rows and reshape
datas      = np.array([[i.split('\t') for i in k] for k in datas]).reshape(ROWCOUNT,COLUMNCOUNT)
#datas_test = np.array([[i.split('\t') for i in k] for k in datas_test]).reshape(7015,COLUMNCOUNT)
#get readeable dataframe value from 'datas'
datas_df      = pd.DataFrame(datas,columns=['Date','Time','Open','High','Low','Close','Tickvol','Vol','Spread'])
#datas_df_test = pd.DataFrame(datas_test,columns=['Date','Time','Open','High','Low','Close','Tickvol','Vol','Spread'])
#String Date + Time value to timestamp value
tmp      = np.array([[datetime.datetime.strptime((datas_df['Date'][i]+" "+datas_df['Time'][i]),'%Y.%m.%d %H:%M:%S').timestamp()] for i in range(ROWCOUNT)]).reshape(ROWCOUNT,1)
#tmp_test = np.array([[datetime.datetime.strptime((datas_df_test['Date'][i]+" "+datas_df_test['Time'][i]),'%Y.%m.%d %H:%M:%S').timestamp()] for i in range(7015)]).reshape(7015,1)
    
date = pd.DataFrame(tmp)
#date_test = pd.DataFrame(tmp_test)
#change values type from string to numeric
for i in datas_df.columns:
    if i !='Date' :
        if i !='Time' :
            datas_df[i] = pd.to_numeric(datas_df[i])
datas_df = datas_df.drop('Date',axis=1)
datas_df = datas_df.drop('Tickvol',axis=1)
datas_df = datas_df.drop('Vol',axis=1)
datas_df = datas_df.drop('Spread',axis=1)
datas_df = datas_df.drop('Time',axis=1)
high= datas_df[['High']]
low= datas_df[['Low']]
close= datas_df[['Close']]
open= datas_df[['Open']]
sec =datas_df

#sec = pd.get_dummies(sec , drop_first=True)

#sec = sec.apply(preprocessing.LabelEncoder().fit_transform)
#date = date.apply(preprocessing.LabelEncoder().fit_transform)
    
#split values for training and testing
from sklearn.model_selection import train_test_split
#x_train_high , x_test_high , y_train_high , y_test_high  = train_test_split(date, high, test_size=0.33, random_state=0,shuffle=True)
#x_train_low  , x_test_low  , y_train_low  , y_test_low   = train_test_split(date, low, test_size=0.33, random_state=0,shuffle=True)
x_train, x_test, y_train, y_test = train_test_split(date, sec, test_size=0.20, random_state=0, shuffle=True)
#n3, x_test, n4, y_test = train_test_split(date_test, sec_test, test_size=1, random_state=0, shuffle=False)
x_train= x_train.values
x_test = x_test.values 
y_train= y_train.values 
y_test = y_test.values
#x_train_open , x_test_open , y_train_open , y_test_open  = train_test_split(date, open, test_size=0.33, random_state=0,shuffle=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=10,random_state=0)

rf_reg.fit(x_train,y_train) #.ravel()

#aaa = [[7190]]
#aaa = np.array(aaa).reshape(1,-1)
    
predict = rf_reg.predict(x_test)
#print(predict)
    
#print(x_test)
x_test = pd.DataFrame(x_test).sort_values([0])
x_test = np.array(x_test)
#print(x_test)
#y_test = pd.DataFrame(y_test).sort_values([0])
#y_test = np.array(y_test)

#plt.plot(x_test[0:50],y_test[0:50],color='red')
#plt.plot(x_test[0:50],predict[0:50])
    
# Calculate the absolute errors
errors = abs(predict - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 5), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print(accuracy)
for i in range(50):
    print(x_test[-i],y_test[-i],predict[-i],(predict[-i]-y_test[-i])*100)
plt.plot(x_test[0:50],y_test[0:50])
plt.plot(x_test[0:50],predict[0:50])
plt.show()
```
