```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.core.fromnumeric import reshape
import pandas as pd
import datetime
import time
from sklearn import preprocessing

ROWCOUNT = 1
COLUMNCOUNT = 9

name = 'M1'
name2 = 'toplu'
name3 = "2008-2022"
#datas      = pd.read_csv('/content/drive/Othercomputers/2022 Masaüstü/Dersler 2022/Dersler-2021-2022-Donem-2/Lisans Projesi/EURUSD_'+name+'.csv', header=1)
datas      = pd.read_csv('C:/Users/ertad/Dersler 2022/Dersler-2021-2022-Donem-2/Lisans Github/EURUSD_'+name+'_'+name3+'.csv', header=1)

datas      = datas.values.tolist()
try :
    datas      = np.array([[i.split('\t') for i in k] for k in datas]).reshape(ROWCOUNT,COLUMNCOUNT)
except Exception as e:
    s = str(e).split()
row = int(int(s[5])/COLUMNCOUNT)

datas      = np.array([[i.split('\t') for i in k] for k in datas]).reshape(row,COLUMNCOUNT)

if COLUMNCOUNT ==8 :
    datas_df   = pd.DataFrame(datas,columns=['Date','Open','High','Low','Close','Tickvol','Vol','Spread'])
elif COLUMNCOUNT ==9:
    datas_df   = pd.DataFrame(datas,columns=['Date','Time','Open','High','Low','Close','Tickvol','Vol','Spread'])
datas_df

#import dill
#dill.dump_session(name+'_'+name2+'-ORJINAL-datas_df.db')
```

```python
tmp =0
if 'Time' in datas_df.columns:
    tmp =pd.to_datetime(datas_df['Date']+" "+datas_df['Time'],format='%Y.%m.%d %H:%M:%S')
    datas_df = datas_df.drop('Time',axis=1)
else :
    tmp =pd.to_datetime(datas_df['Date'],format='%Y.%m.%d')
    
#datas_df = datas_df.drop(['Date'],axis=1)
datas_df = datas_df.drop(columns=['Vol'])
datas_df = datas_df.drop(columns=['Tickvol'])
datas_df = datas_df.drop(columns=['Spread'])
tmp
```

```python
date2 = pd.DataFrame()
date2['Date'] = tmp
date = tmp
newDate = pd.DatetimeIndex(date)

n2 = pd.DataFrame()
m  = (2 * np.pi * newDate.month/12)
w  = (2 * np.pi * pd.Series(newDate).dt.isocalendar().week/(52.177457))
d  = (2 * np.pi * newDate.day/(365.242199))
h  = (2 * np.pi * newDate.hour/24)
mi = (2 * np.pi * newDate.minute/60)

n2['month_sin']  =  np.round(np.sin(m),5)
n2['month_cos']  =  np.round(np.cos(m),5)
n2['weekly_sin'] =  np.round(np.sin(w),5)
n2['weekly_cos'] =  np.round(np.cos(w),5)
n2['day_sin']   =  np.round(np.sin(d),5)
n2['day_cos']   =  np.round(np.cos(d),5)
n2['hour_sin']  =  np.round(np.sin(h),5)
n2['hour_cos']  =  np.round(np.cos(h),5)
n2['minute_sin'] = np.round(np.sin(mi),5)
n2['minute_cos'] = np.round(np.cos(mi),5)

n2['sin'] = n2['month_sin'] + n2['weekly_sin'] + n2['day_sin'] + n2['hour_sin'] + n2['minute_sin']
n2['cos'] = n2['month_cos'] + n2['weekly_cos'] + n2['day_cos'] + n2['hour_cos'] + n2['minute_cos']

datas_df = datas_df.drop(columns=['Date'])
#datas_df = datas_df.drop(columns=['Time'])

datas_df = pd.concat([
                      #date2['Date'],
                      datas_df,
                      n2['sin'],n2['cos'] 
                      ],axis=1)
datas_df
```
